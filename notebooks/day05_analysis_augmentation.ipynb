## Observations
*Focus: measurable results only*

The SSIM evaluation shows that gamma correction has a moderate impact on structural similarity in the Sentinel-2 satellite composite, with SSIM decreasing as gamma deviates from the baseline value of 1.0 in either direction. The strongest gamma corrections ($\gamma = 0.5$ and $\gamma = 2.0$) produced the largest SSIM drops, indicating significant structural distortion from extreme brightness changes. For Gaussian noise, SSIM decreases monotonically with increasing $\sigma$. Even mild noise ($\sigma = 0.02$) causes a measurable SSIM reduction, while severe noise ($\sigma = 0.10$) leads to a substantial drop, indicating significant structural degradation. Performance differences are most pronounced in the noise sweep, where SSIM drops more steeply than for moderate gamma variations.

## Interpretation
*Focus: explain observed effects based on methods and parameters*

The observed SSIM decrease under gamma correction is caused by the non-linear intensity remapping, which alters the luminance and contrast components of the SSIM metric. At $\gamma = 0.5$, bright regions (snow/firn in the Sentinel-2 composite) become saturated and lose detail, compressing the local contrast at ice-rock boundaries. At $\gamma = 2.0$, dark regions (rock faces and moraine deposits) compress toward zero, similarly reducing boundary visibility. This indicates that gamma correction directly affects the contrast gradient at structural transitions, which is critical for distinguishing ice from rock in the satellite-based glacier monitoring application.

For noise injection, the SSIM decrease is driven by the corruption of local variance and covariance patterns that SSIM measures. At the 10 m pixel resolution of Sentinel-2, each pixel integrates a relatively large ground area, so the per-pixel noise effect is particularly disruptive to fine boundary structures. Higher $\sigma$ introduces random intensity fluctuations that obscure the subtle spectral differences between adjacent ice and rock pixels, while the overall image structure (large-scale glacier shape) remains recognizable even at $\sigma = 0.10$.

## Discussion and Critical Reflection
*Focus: practical applicability and limitations*

For the defined satellite-based glacier monitoring use case, moderate gamma corrections ($\gamma \in [0.75, 1.5]$) and low noise levels ($\sigma \leq 0.02$) are the most suitable augmentation configurations because they maintain SSIM values close to the baseline, preserving the ice-rock boundary visibility required for glacier extent delineation from Sentinel-2 imagery. In contrast, extreme gamma values ($\gamma = 0.5$, $\gamma = 2.0$) and high noise levels ($\sigma \geq 0.05$) produce configurations where structural features are significantly degraded.

The achieved SSIM levels for moderate augmentations suggest they could be safely used to expand training datasets for automated glacier boundary detection without introducing misleading structural artifacts. However, SSIM is a global metric that does not specifically target the ice-rock boundary — a region-specific evaluation (e.g., SSIM computed only over the ROI at the glacier margin) could provide more targeted assessment.

A limitation specific to satellite imagery is that Sentinel-2 surface reflectance values are already radiometrically corrected, meaning the baseline image has undergone atmospheric correction. The augmentations applied here therefore simulate degradations on top of a corrected product, which may not perfectly match the characteristics of raw, uncorrected acquisitions. Future work could address this by applying augmentations to Level-1C (top-of-atmosphere) products instead, or by evaluating augmentation effects across multiple temporal composites to assess seasonal robustness.
