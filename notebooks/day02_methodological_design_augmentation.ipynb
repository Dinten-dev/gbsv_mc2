## Theoretical Foundation and Method Choice
*Focus: principled justification aligned with the use case*

This investigation applies the principle of **intensity transformation** and **noise modeling** to Sentinel-2 satellite imagery of the Aletsch Glacier within the context of satellite-based glacier monitoring in Switzerland. Non-geometric augmentation alters pixel intensities without modifying spatial structure, simulating real-world imaging variations such as changing atmospheric conditions or sensor noise. Two methods are selected:

1. **Gamma correction** — a power-law transformation $I_{out} = I_{in}^\gamma$ that non-linearly adjusts image brightness. This models varying atmospheric transparency and illumination conditions encountered in satellite imagery of alpine terrain. For $\gamma < 1$, the image brightens (simulating high atmospheric transparency or strong snow reflections); for $\gamma > 1$, the image darkens (simulating haze, thin cloud cover, or terrain shadows).

2. **Additive Gaussian noise** — pixel-wise addition of $\mathcal{N}(0, \sigma^2)$ noise, simulating sensor noise in the Sentinel-2 Multi-Spectral Instrument (MSI). This is relevant because satellite sensors at orbital altitude are subject to quantization noise, atmospheric scattering effects, and radiometric calibration uncertainties.

These methods are selected because they independently target two fundamental vulnerabilities in satellite-based glacier monitoring: brightness variation due to atmospheric conditions and noise degradation from sensor limitations. If brightness is distorted beyond a critical point, the ice-rock contrast collapses. If noise is too high, fine structural details at the 10 m pixel scale are lost.

## Parameter Definition and Mathematical Specification
*Focus: explicit parameter selection, derivation, and unit consistency*

The image is a 3-channel RGB image derived from Sentinel-2 bands B4, B3, B2 with pixel intensities rescaled to $[0, 255]$ (8-bit unsigned integer). For processing, intensities are normalized to $[0, 1]$.

**Gamma correction parameters:**
- The transformation is defined as $I_{out}(x,y) = I_{in}(x,y)^\gamma$, applied per channel
- $\gamma = 1.0$ corresponds to the identity (baseline, no change)
- $\gamma < 1.0$: brightening (e.g., $\gamma = 0.5$ roughly doubles perceived brightness)
- $\gamma > 1.0$: darkening (e.g., $\gamma = 2.0$ significantly reduces brightness in mid-tones)
- Selected values: $\gamma \in \{0.5, 0.75, 1.0, 1.5, 2.0\}$

**Gaussian noise parameters:**
- Noise is drawn from $\mathcal{N}(0, \sigma^2)$ and added to the normalized image
- $\sigma = 0$: no noise (baseline)
- $\sigma$ is expressed in normalized intensity units $[0, 1]$
- Selected values: $\sigma \in \{0, 0.02, 0.05, 0.10\}$
- $\sigma = 0.02$ represents mild sensor noise typical of Sentinel-2, $\sigma = 0.10$ represents severe degradation beyond normal operating conditions

In the satellite-based glacier monitoring domain, $\gamma$ maps to atmospheric transparency and illumination conditions, while $\sigma$ maps to sensor quality and radiometric calibration uncertainty. These parameter choices span realistic to extreme conditions.

## Experimental Design for Next Days
*Focus: structured parameter variation and theoretical prediction*

The baseline configuration is defined as $\gamma = 1.0$ with $\sigma = 0$ (original unmodified satellite image). Two independent parameter sweeps will be conducted:

**Sweep 1 — Gamma variation (noise fixed at $\sigma = 0$):**
$\gamma \in \{0.5, 0.75, 1.0, 1.5, 2.0\}$, yielding 5 configurations.

**Sweep 2 — Noise variation (gamma fixed at $\gamma = 1.0$):**
$\sigma \in \{0, 0.02, 0.05, 0.10\}$, yielding 4 configurations.

This gives a total of **8 unique configurations** (baseline counted once).

It is theoretically expected that:
- Decreasing $\gamma$ below 1.0 will brighten the image, potentially saturating snow/ice regions and reducing ice-rock contrast
- Increasing $\gamma$ above 1.0 will darken the image, potentially losing detail in shadowed rock areas and dark moraine deposits
- Increasing $\sigma$ will progressively degrade fine structural features at the 10 m pixel scale, such as boundary sharpness between ice and rock, while large-scale features (overall glacier shape) should remain visible longer

All configurations will be documented with explicit parameter labeling to ensure traceability.

## Methodological Limitations and Risk Factors
*Focus: assumptions, stability, and potential misinterpretation*

This approach assumes that gamma correction and additive Gaussian noise are adequate models for real-world imaging variations in satellite-based glacier observation. In practice, atmospheric degradation is spatially non-uniform (e.g., partial cloud shadows, variable aerosol optical depth), which gamma correction — being a global operation — cannot capture. Similarly, real sensor noise in Sentinel-2 may exhibit band-dependent characteristics or non-Gaussian distributions (e.g., striping artifacts).

The method is expected to be reliable when the image has sufficient dynamic range and the structures of interest (ice-rock boundaries) exhibit clear contrast. However, it may become misleading when the original image already contains saturated regions (snow/ice at intensity 255), since gamma correction cannot recover clipped information. In this glacier monitoring use case, the primary risk is that snow-covered areas in the Sentinel-2 composite may already be near saturation, limiting the effective range of brightening augmentations.
